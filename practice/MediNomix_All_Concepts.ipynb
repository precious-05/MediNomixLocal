### **Backend FastApi**

```python
from fastapi import FastAPI

app = FastAPI()
@app.get("/hello/{name}")
def sayHello(name: str):
    print("Backend function executed successfully!")
    return {"message" : f"Hello {name}"}
#Here we are returning the response to the get request

    
```

- / path define karta hai — ye URL ka route hota hai, jaise /api/search/{drug_name}.
- {drug_name} path parameter hota hai — jo frontend URL me bhejta hai, wo yahan receive hota hai.
- Function arguments me wahi cheez aati hai jo frontend se request ke through aati hai (path, query, body).
- drug_name: str me str type hint hoti hai — FastAPI automatically type validate karta hai.
- Depends() wale arguments frontend se nahi aate — wo backend internally inject karta hai (jaise DB session).

### **Model Response**
- **`The BaseModel is a class from the Pydantic library used to define the structure and validation requirements for your data. It acts as the core mechanism for handling data in your API, ensuring type safety, providing automatic validation, and generating interactive API documentation`**
- **`Pydantic is a popular Python library for data validation, serialization, and settings management, using standard Python type hints to ensure data conforms to expected types. It automatically parses, coerces (e.g., "1" to 1), and validates data, raising errors if input is invalid`**
- **`In Pydantic, serialization is the process of converting a structured Pydantic model instance into less structured, standard Python data types (like a dictionary) or a JSON-encoded string. This is essential for tasks such as sending data over a network via an API, storing data in a database, or logging information`**

```python
from fastapi import FastAPI
from pydantic import BaseModel
app=FastAPI()

class SumResponse(BaseModel):
    OriginalNumber: int
    Sum: int
    
@app.get("/sum/{num}", response_model=SumResponse)
def CalculateSum(num: int):
    print(f"Running the square function in backend with request value of {num}")
    return{
        "OriginalNumber" : num,
        "Sum" : num + num,
        "extra_feild" : "ignored"
    }
    
```

- Haan, response_model sirf wahi fields allow karta hai jo SumResponse class me defined hain.
- Jo extra fields return me hongi (jaise "extra_field"), FastAPI unhe automatically ignore kar dega.
- Pydantic field names exact match hone chahiye (number, sum), warna validation error aa sakta hai.
- Isliye best practice: return keys ka naam model ke variables jaisa hi rakhy jyen

### **Dependency Injection**
- Dependency Injection (DI) ka simple matlab hai:
- ek function ko jo cheez chahiye (dependency), wo khud create na kare, balki bahar se usko de di jaye
- code clean or modular ho jata ha
- Reusability badh jati hai (same DB connection multiple endpoints me)
- FastAPI Dependency injection khud manage krta ha

```python
from fastapi import FastAPI, Depends
app = FastAPI()


def get_multiplier():
    print("Called Multiplier Function and providing dependency to multiply function")
    return 3

@app.get("/multiply/{num}")
def multiply(num:int, multiplier: int = Depends(get_multiplier)):
    print("Running Multiply Function")
    result = num * multiplier
    return{
        "number" : num,
        "multiplier": multiplier,
        "result": result
    }
```
