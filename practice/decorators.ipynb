### **Decorators**

````python
def changecase(function1):   #changecase is the decorator
    def inner():
        return function1().upper()
    return inner


@changecase    #function1 is being decorated
def function1():
    return "Hello Alina"


print(function1())
````

````python
def changecase(greet):
    def inner():
        return greet().lower()
    return inner



@changecase
def greet():
    return "Alina"


print(greet())
````

Every variable has three important things:

* a **name**
* a **type**
* a **value**

For example:

```python
earth_weight = 120
```

Here:

* the name is `earth_weight`
* the value is `120`

Another important concept is **type casting**.

When we use `input()` in Python, it always returns a **string**, even if the user types a number. So if we want to do mathematical calculations, we need to convert that string into a number using functions like `float()` or `int()`.

```python
def main():
```
